```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv('/content/student scores.csv')
df.head()
```

```python
X=df.drop('Scores',axis=1)
y=df['Scores']
y
```

```python
df.shape
```

```python
plt.figure(figsize=(4,4))
plt.scatter(X,y)
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=71)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
aplpha=model.intercept_
beta=model.coef_
print(aplpha)
print("beta",beta)
```

```python
y_pred_train=model.predict(X_train)
y_pred_test=model.predict(X_test)
```

```python
plt.figure(figsize=(7,3))
# subplot
plt.subplot(1,2,1)
plt.title('Training case')
plt.scatter(y_train,y_pred_train)
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.grid(True)


# subplot 2
plt.subplot(1,2,2)
plt.scatter(y_test,y_pred_test)
plt.grid(True)
plt.title('Testing case')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
mse_train=mean_squared_error(y_train,y_pred_train)
mae_train=mean_absolute_error(y_train,y_pred_train)

mse_test=mean_squared_error(y_test,y_pred_test)
mae_test=mean_absolute_error(y_test,y_pred_test)
```

```python
# create a dataframe of the results

performance_df=pd.DataFrame({
    'Metric':['MSE','MAE','MSE','MAE'],
    'DataType':['train','train','test','test'],
    'Values':[mse_train,mae_train,mse_test,mae_test]},
)
performance_df.T
```
