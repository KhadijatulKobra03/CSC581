```python
import pandas as pd
# create a data frame

data ={
    'Name':['Alice','Bob','Charlie','David'],
    'Age':[25,30,17,28],
    'City':['New York','San Francisco','New Haven','Los Angeles'],
    'Salary':[100,150,200,300]
}
df = pd.DataFrame(data)
df
```

```python
# calculate some statistics on each column
df['Age'].mean()
```

```python
print("The minumum salary is",df['Salary'].min())
print("The maximum salary is",df['Salary'].max())
print("The mean salary is",df['Salary'].mean())
```

```python
ndata =[10,20,30,40]
s= pd.Series(ndata)
print(s.min())
print(s.max())
print(s.mean())
```
