```python
import matplotlib.pyplot as plt
x =[1,2,3,4,5]
y =[10,15,7,12,8]


#create a line plot
plt.plot(x,y, marker='o',markersize=6,color='blue', linestyle='--',linewidth=2)

#add labels and titiles
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('X versus Y')
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Create a DataFrame

data={
    'Year':[2010,2011,2012,2013,2014],
    'Sales':[500,800,1700,1500,2000]
}
df=pd.DataFrame(data)

# Create a line plot
plt.plot(df['Year'],df['Sales'],marker='o')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Over Years')
plt.grid(True)

# Set x-axix ticks at integer values
plt.xticks(df['Year'])

plt.show()
```

Bar Plot

```python
data={
    'Category':['A','B','C','D'],
    'Sales':[500,800,1200,1500]
}
df=pd.DataFrame(data)
# Create a bar plot
plt.bar(df['Category'],df['Sales'],color='green')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Sales by Category')

plt.show()
```

```python
data={
    'Age':[25,30,32,26,28],
    'Income':[50000,60000,90000,8500,70000]
}
df=pd.DataFrame(data)
# Create a scatter plot
plt.scatter(df['Age'],df['Income'],color='red')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Age vs Income')
plt.grid(True)

plt.show()
```

```python
# Creating a data frame
data={
    'Scores':[85,92,78,90,88,76,95,82,70],
}
df=pd.DataFrame(data)

# Creating a histogram of the data
plt.hist(df['Scores'],width=1.5)
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('distribution of Scores')
plt.grid(True)

plt.show()
```
